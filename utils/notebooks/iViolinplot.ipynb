```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
```

```python
fig=go.Figure()
df=pd.read_csv("Datasets/tips.csv")
fig.add_trace(go.Box(x=df['day'],y=df["total_bill"],marker_color=["red","yellow","green","blue"]))
```

```python
tmp=df.copy()
colors=["gold","orange","olive","salmon"]
unique_vals=tmp["day"].unique()
dummie=dict()
for col,val in zip(colors,unique_vals):
    dummie[val]=col
tmp['color_col']= tmp['day'].map(dummie)
```

```python
for each,color in zip(df['day'].unique(),['green', 'red', 'yellow',"olive"]):
    fig.add_trace(go.Box(name=each,y=df["total_bill"],fillcolor=color))
fig.show()
```
